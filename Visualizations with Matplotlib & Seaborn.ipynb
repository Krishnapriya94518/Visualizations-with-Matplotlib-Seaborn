```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df1 = pd.read_csv(r"C:\Users\MY_Home\Downloads\archive\application_record.csv") 
df2 = pd.read_csv(r"C:\Users\MY_Home\Downloads\archive\credit_record.csv")

df = pd.merge(df1, df2, on="ID", how="outer")  


# Task 1:
df = df[df['DAYS_EMPLOYED'] < 0] 
df['DAYS_EMPLOYED'] = df['DAYS_EMPLOYED'].abs()
df['DAYS_BIRTH'] = df['DAYS_BIRTH'].abs()

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='DAYS_BIRTH', y='DAYS_EMPLOYED', alpha=0.5, linewidth=0.5)
plt.xlabel("Age in Days")
plt.ylabel("Days Employed")
plt.title("Scatterplot of Days Employed vs Age")
plt.show()

#Task 2:
df['AGE_YEARS'] = df['DAYS_BIRTH'] // 365

plt.figure(figsize=(10, 6))
sns.histplot(df['AGE_YEARS'], kde=True, bins=30)
plt.xlabel("Age in Years")
plt.ylabel("Density")
plt.title("Age Distribution of Individuals")
plt.show()

# Task 3:
median_income = df['AMT_INCOME_TOTAL'].median()
bottom_half_df = df[df['AMT_INCOME_TOTAL'] <= median_income]

plt.figure(figsize=(10, 6))
sns.boxplot(data=bottom_half_df, x='NAME_FAMILY_STATUS', y='AMT_INCOME_TOTAL', order=['Single / not married', 'Married', 'Widow', 'Civil marriage', 'Separated'])
plt.xticks(rotation=45)
plt.xlabel("Family Status")
plt.ylabel("Total Income")
plt.title("Income Distribution by Family Status (Bottom 50% Earners)")
plt.show()
```
